```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
# make_blobs help in generate the cluster datasets
x, y = make_blobs(n_samples=1000, centers = 3, n_features=2, random_state=1 )
```

```python
x
```

```python
y.shape
```

```python
# In unsupervised learning you don't use y (D.V)
```

```python
plt.scatter(x[:, 0], x[:, 1])
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans
```

```python
kmeans.fit(x_train)  # fit_predict can also be used
```

```python
y_labels = kmeans.predict(x_train)
y_labels
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1], c = y_labels)
```

```python
# how better this clustring model is ?

from sklearn.metrics import silhouette_score
silhouette_score(x_train, kmeans.labels_)  # close to 1, we have good model
```

```python
kmeans.labels_
```

```python
kmeans.inertia_  # k =3, this is wcsss distance
```

```python
# how to select k(centroid) >> elbow method
```

```python
wcss = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss  # wcss is decreasing as the k value is increasing
```

```python
# plot the elbow curve

plt.plot(range(1, 10), wcss)
plt.xticks(range(1, 10))
plt.xlabel(" No. of students")
plt.ylabel("wcss")
plt.show()
```

```python
# based on the plot above the optimal no. of k/centroid should be 3
```
