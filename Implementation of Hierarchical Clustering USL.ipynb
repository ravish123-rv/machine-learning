# Implementation of Hierarchical Clustering

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
data = datasets.load_iris()
x = data.data
y = data.target
```

```python
x
```

```python
y.shape
```

```python
# this is Unsupervised Learning algorithm so y is not use
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
# applying the PCA

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_scaled = pca.fit_transform(x_scaled)
pca_scaled
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c = y)
```

```python
# agglomerative clustring >> make dendogram

import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20, 7))
plt.title("Dendogram")

sc.dendrogram(sc.linkage(pca_scaled, method = 'ward'))

plt.xlabel("Sample data")
plt.ylabel("Euclidian distance")
plt.show()
```

```python
# Build the model

from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
cluster
```

```python
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c = cluster.labels_ )
```

```python
from sklearn.metrics import silhouette_score

silhouette_coef = []
for k in range(2, 11):
    agglo = AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage = 'ward')
    agglo.fit(x_scaled)
    score = silhouette_score(x_scaled, agglo.labels_)
    silhouette_coef.append(score)
```
