```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the data (ingestion of data)
# we are loading the data from scikit learn
# scikit learn used in ML in python >>
# 1. Simple and efficient tools for predictive data analysis
# 2. Accessible to everybody, and reusable in various contexts
# 3. Built on NumPy, SciPy, and matplotlib
# 4. Open source, commercially usable - BSD license
# in scikit learn libraray >> sklearn is a moduel of scikit learn to give data set and preprocessing the data

from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
diabetes
```

```python
# Problem statement : To pridict progression one year after baseline provided10 features
# in one year how the suger level has changed
```

```python
data = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
data
```

```python
data.head()
```

```python
data['target'] = diabetes.target
```

```python
data
```

```python
# After these done we do the EDA , DATACLEANING, DATA PREPERATION, FEATURE ENGNEERING
```

```python
data
```

```python
# in SLR one x and one y 
x = data[['bmi']] # one independent variable 
y = data['target'] # one dependent variable
```

```python
x
```

```python
y
```

```python
# using sklearn we are done the train-test split the data
# using train_test_split the date are split and train and test the model
#  train_test_split are split the data in random manner
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)

# random_state when fixed hepls us to reproduce the same result again & again
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# scalling (optional)
```

```python
# Model Traning
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
# orange color indicate that model is not trained
```

```python
x_train
```

```python
y_train
```

```python
# Training of model
model.fit(x_train, y_train)
# At the backend it will use gradient descent to give optimal coefficient 
# blue color is indicate that model is trained
```

```python
model.coef_  # attribute of the model
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# visualise the result
plt.scatter(x_test, y_test, color = 'black', label = 'Actual')
plt.plot(x_test, y_pred, color = 'blue', linewidth = 3, label = "predicated linear regression" )
plt.xlabel("BMI")
plt.ylabel("one year progression target")
plt.title("Linear Regression on Diabetes data")
plt.legend()
plt.show()
```

```python
# model evalution

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
