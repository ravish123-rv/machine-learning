# Ensemble Technique

# Bagging Regressor or Hard Voting Regressor

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
X, y = make_regression(n_samples = 100, n_features = 10, noise = 0.1, random_state = 1 )
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
# for model creation
 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
# for model creation
lr = LinearRegression()
dtr = DecisionTreeRegressor()
svr = SVR(kernel = 'linear')
```

```python
# to create  esembles

from sklearn.ensemble import VotingRegressor
ensemble_regressor = VotingRegressor(estimators = [('linear_reg', lr), ('DTR', dtr), ('SVR', svr)])
```

```python
ensemble_regressor
```

```python
# model traing

ensemble_regressor.fit(X_train, y_train)
```

```python
# model prediction

y_pred = ensemble_regressor.predict(X_test)  # avg. of all the prediction will be final prediction
y_pred
```

```python
# model evalution 

from sklearn.metrics import r2_score
r2_score(y_pred, y_test)
```
