```python
# since we have not done outlier treatment , then better idea would be to 
# replace the missing value with median

df['horsepower'].median()
```

```python
# origin is the categrical column so we convert into numerical column 
# this process is known as Data Encoding
```

```python
# DAta Encoding >> done with help of using dictionary
# how to map the dictionary with df['origin'] >> using map function
df['origin'] = df['origin'].map({"usa" :1, "japan" :2, "europe" :3})
```

```python
#df['origin'] = df['origin'].astype(int) # convert into integer data type
```

```python
# sepertate the data into x(I.V) & y(D.V) 
```

```python
X = df.drop('mpg', axis = 1)
y = df['mpg']
```

```python
X
```

```python
y
```

```python
# train test split with X & y

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 1)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# crestion of simple Linear regression model

from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
```

```python
regression_model
```

```python
regression_model.fit(X_train, y_train)
```

```python
regression_model.coef_
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {regression_model.coef_[i]}")
    
```

```python
# observations >> cofficients are relatively smaller, if one independent
# variable changes there will be not much difference in prediction
# this is sometime called as smoother model

# these features might not be contributing in model training
```

```python
from sklearn.metrics import r2_score

y_pred_linear = regression_model.predict(X_test)
r2_linear = r2_score(y_test, y_pred_linear)
print(f"The R square of linear regression {r2_linear}")
```

# Regularized model

```python
# Ridge regression 
# in practicle implementation lambda is alpha

from sklearn.linear_model import Ridge
ridge_regression_model = Ridge(alpha = 0.1)
ridge_regression_model
```

```python
ridge_regression_model.fit(X_train, y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {ridge_regression_model.coef_[i]}")
```

```python
# for ridge regression evelution 

y_pred_ridge = ridge_regression_model.predict(X_test)
r2_ridge = r2_score(y_test, y_pred_ridge)
print(f"The R square of ridge regression {r2_ridge}")
```

```python
# observation : we dont see much variation in coeff of ridge regression as comared to linear regression 
```
