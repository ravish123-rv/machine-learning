```python
# origin is the categrical column so we convert into numerical column 
# this process is known as Data Encoding
```

```python
# DAta Encoding >> done with help of using dictionary
# how to map the dictionary with df['origin'] >> using map function
df['origin'] = df['origin'].map({"usa" :1, "japan" :2, "europe" :3})
```

```python
X = df.drop('mpg', axis = 1)
y = df['mpg']
```

```python
X
```

```python
y
```

```python
# train test split with X & y

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 1)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# crestion of simple Linear regression model

from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
```

```python
regression_model
```

```python
regression_model.fit(X_train, y_train)
```

```python
regression_model.coef_
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {regression_model.coef_[i]}")
    
```

```python
# observations >> cofficients are relatively smaller, if one independent
# variable changes there will be not much difference in prediction
# this is sometime called as smoother model

# these features might not be contributing in model training
```

```python
from sklearn.metrics import r2_score

y_pred_linear = regression_model.predict(X_test)
r2_linear = r2_score(y_test, y_pred_linear)
print(f"The R square of linear regression {r2_linear}")
```

# Regularized model

```python
# Ridge regression 
# in practicle implementation lambda is alpha

from sklearn.linear_model import Ridge
ridge_regression_model = Ridge(alpha = 0.1)
ridge_regression_model
```

```python
ridge_regression_model.fit(X_train, y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {ridge_regression_model.coef_[i]}")
```

```python
# for ridge regression evelution 

y_pred_ridge = ridge_regression_model.predict(X_test)
r2_ridge = r2_score(y_test, y_pred_ridge)
print(f"The R square of ridge regression {r2_ridge}")
```

```python
# observation : we dont see much variation in coeff of ridge regression as comared to linear regression 
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import Lasso, Ridge
```

```python
# grid search cv for Lasso regularisation
# define the model
lasso = Lasso()
lasso
```

```python
# grid search cv >> Dictionary with parameters names(str) as keys and Lists of 
# parameter settings to try as value

param_grid = {'alpha' : [0.001, 0.01, 0.1, 1, 10, 100]}
param_grid
```

```python
grid_search = GridSearchCV(estimator=lasso, param_grid=param_grid, cv = 5, scoring = 'r2', verbose = 2)
```

```python
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_params_['alpha']
```

```python
grid_search.best_score_  # possible avg. score here
```

```python
grid_search.best_estimator_  # estimator >> model
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
r2_score(y_test, y_pred)
```

```python
# random search cv for Lasso
lasso = Lasso()
param_distributions = {'alpha' : [0.001, 0.01, 0.1, 10, 1, 100]}
random_Search_lasso = RandomizedSearchCV(estimator = lasso, param_distributions = param_distributions, cv = 5, n_iter=10, scoring = 'r2',verbose = 2 )
```

```python
random_Search_lasso
```

```python
random_Search_lasso.fit(X_train, y_train)

# due to n_iter parameter in each cross validation random 3 parameter are selected
```

```python
random_Search_lasso.best_estimator_
```

```python
random_Search_lasso.best_score_
```

```python
y_pred = random_Search_lasso.best_estimator_.predict(X_test)
r2_score(y_test, y_pred)
```

```python
# elastic net Grid Search CV

elastic = ElasticNet()
param_grid = {'alpha' : [0.001, 0.01, 0.1, 1, 10, 100],
              'l1_ratio' : [0.1, 0.4, 0.9]}
elastic = GridSearchCV(estimator = elastic, param_grid = param_grid, cv = 5, scoring = 'r2', verbose = 2)
elastic.fit(X_train, y_train)
print(elastic.best_params_)

print(elastic.best_estimator_)

y_pred = elastic.best_estimator_.predict(X_test)

print(r2_score(y_test, y_pred))
```

```python
# elastic net Randomize Search CV

elastic = ElasticNet()
param_distributions = {'alpha' : [0.001, 0.01, 0.1, 1, 10, 100],
              'l1_ratio' : [0.1, 0.4, 0.9]}
elastic = RandomizedSearchCV(estimator = elastic, param_distributions = param_distributions, n_iter = 3, cv = 5, scoring = 'r2', verbose = 2)
elastic.fit(X_train, y_train)
print(elastic.best_params_)

print(elastic.best_estimator_)

y_pred = elastic.best_estimator_.predict(X_test)

print(r2_score(y_test, y_pred))
```
