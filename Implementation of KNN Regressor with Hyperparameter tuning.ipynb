# KNN - Regressor

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression

x, y = make_regression(n_samples=1000, n_features=3, noise=10, random_state=1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

# split the data int x and y datasets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

```python
x_train.shape, x_test.shape
```

```python
# creation of model
from sklearn.neighbors import KNeighborsRegressor

reg =  KNeighborsRegressor()
```

```python
reg
```

```python
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print(r2_score(y_test, y_pred))
```

```python
print(mean_absolute_error(y_test, y_pred))
```

```python
print(mean_squared_error(y_test, y_pred))
```

```python
# hyperparameter tuning

from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_neighbors' : [3, 5, 6, 7, 9, 11, 13],
    'weights' : ['uniform', 'distance'],
    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size' : [20, 30, 40, 50]
}
```

```python
grid_search = GridSearchCV(estimator=reg,param_grid=param_grid, cv=5, verbose=3 )
grid_search
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```
